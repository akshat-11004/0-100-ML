```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)
```

```python
df = pd.read_excel('Real_estate.xlsx')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
df_duplicated=df[df.duplicated()]
df_duplicated.shape
df.count()
```

```python
df = df.drop_duplicates()
df.count()
```

```python
df.dtypes
```

```python
sns.boxplot(x=df['Y house price of unit area'])
```

```python
Q1= df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3-Q1
print(IQR)
```

```python
df = df[~((df<Q1-1.5*IQR) |(df>Q3+1.5*IQR))]
df.shape
```

```python
plt.figure()
c=df.corr()
sns.heatmap(c,cmap = "BrBG",annot = True)
```

```python
plt.figure(figsize=(10, 5))  # Bar graph
plt.bar(df['X2 house age'], df['Y house price of unit area'], color='blue',width = 1)
plt.xlabel('X2 house age')
plt.ylabel('Y house price of unit area')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.scatter(df['X2 house age'], df['Y house price of unit area'])
ax.set_xlabel('X2 house age')
ax.set_ylabel('Y house price of unit area')
plt.show()
```
